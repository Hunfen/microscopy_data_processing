```python
from skimage import io
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
from math import sqrt
rcParams['mathtext.fontset'] = 'stix'
```

```python
coordinates = np.genfromtxt('/Users/hunfen/OneDrive/General Files/ゼミー/20201030/20201027FeSeDepo/cOordinates.txt', delimiter = ',').astype(int)
IntensityLog = np.zeros(239)
rect_size = (coordinates[0] - coordinates[2]) * (coordinates[1] - coordinates[3])

for i in range(239):
    fig_data = io.imread('/Users/hunfen/OneDrive/General Files/ゼミー/20201030/20201027FeSeDepo/Image' + str(i + 1).zfill(4) + '.jpg')

    for y in range(coordinates[1], coordinates[3]):
        for x in range(coordinates[0], coordinates[2]):
            IntensityLog[i] += fig_data[y][x]
    IntensityLog[i] = IntensityLog[i] / rect_size
```

```python
fig, axis = plt.subplots()
axis.plot(np.linspace(0, 480, 239), IntensityLog, '.', color = 'black')
axis.set_title(r'$\bf{Inetnsity\ Log}$', fontsize =18)
axis.set_xlabel(r'$t\ \mathrm{[s]}$', fontsize = 14)
axis.set_ylabel(r'$Intensity\ \mathrm{[a.u.]}$', fontsize = 14)
axis.set_yticks([])

axis.axvline(x = 132, ls = '--', color = 'green')
axis.axvline(x = 424, ls = '-', color = 'red')
axis.text(325, 160,r'$3\mathrm{min}43\mathrm{sec}$', color = 'black', fontsize = 12)
axis.arrow(355, 155, 0, -20, width = 3, color = 'black')
axis.text(92, 125, '$132\ \mathrm{s}$', fontsize =12)
axis.arrow(90, 120, 30, 0, width = 1.5, color = 'green')
axis.text(382, 125, '$424\ \mathrm{s}$', fontsize =12)
axis.arrow(380, 120, 30, 0, width = 1.5, color = 'red')
# 
fig.tight_layout();
fig.savefig('/Users/hunfen/OneDrive/General Files/ゼミー/20201030/IntensityLog.jpg', dpi = 600)
```
