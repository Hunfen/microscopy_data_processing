```python
spec_1UC_120pm_FeSe_50mV = np.genfromtxt('/Users/hunfen/OneDrive/General Files/ゼミー/20201009/spec_1UC_120pm_FeSe_50mV.txt', delimiter = ';')
print(spec_1UC_120pm_FeSe_50mV.shape)
```

```python
# average
j = 0
spec_1UC_120pm_FeSe_50mV_avg = np.zeros(329)
for i in range(0, 127):
    if i % 2 == 1:
        spec_1UC_120pm_FeSe_50mV_avg += spec_1UC_120pm_FeSe_50mV[:, i]
        j += 1
spec_1UC_120pm_FeSe_50mV_avg = spec_1UC_120pm_FeSe_50mV_avg / j
axis.plot(spec_1UC_120pm_FeSe_50mV[:, 0], spec_1UC_FeSe_50mV_avg, ls = '-', lw = 3, color = 'black', label = '1UC')
spec_1UC_120pm_FeSe_50mV[:, 0] = (spec_1UC_120pm_FeSe_50mV[:, 0] - 0.005) * 1e3
```

```python
fig, axis = plt.subplots(figsize = (5, 5))
axis.plot(spec_1UC_120pm_FeSe_50mV[30:, 0], spec_1UC_120pm_FeSe_50mV_avg[30:], ls = '-', lw = 3, color = 'black', label = '1UC + 120pm')
axis.set_xlabel('Bias(mV)', fontsize = 18)
axis.set_ylabel('dI/dV(a.u.)', fontsize = 18)
axis.axhline(y = 0, color = 'black', lw = 1);
fig.tight_layout();
```

```python
fig, axis = plt.subplots(figsize = (5, 5))

axis.axhline(y = 0, color = 'black', lw = 1)

# spec plot 1UC
axis.plot(spec_1UC_FeSe_50mV[20:, 0], spec_1UC_FeSe_50mV[20:, 127], ls = '-', lw = 3, label = '1UC', color = 'black')
axis.set_xlabel('Bias(mV)', fontsize = 18)
axis.set_ylabel('dI/dV(a.u.)', fontsize = 18)
axis.axhline(y = 0, color = 'black', lw = 1)

# spec plot 2UC
axis.plot(spec_1UC_120pm_FeSe_50mV[30:, 0], spec_1UC_120pm_FeSe_50mV_avg[30:] + 0.2, label = '1UC + 120pm', ls = '-', lw = 3, color = 'red')

#
axis.axvline(x = 8, color = 'black', ls = '--',lw = 1)
axis.axvline(x = -9, color = 'black', ls = '--',lw = 1)

# legend
axis.legend()

axis.set_title(' ', fontsize = 18)

fig.tight_layout();
```
