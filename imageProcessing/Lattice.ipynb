```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def mirror(a, b, c, p = np.array([0, 0])):
    temp = -2 * (a * p[0] + b * p[1] + c) /(a * a + b * b)
    return np.array([temp * a + p[0], temp * b + p[1]])
```

```python
def rot(a, b, c, d, angle):
    return np.array([(a - c) * np.cos(angle) - (b - d) * np.sin(angle) + c, (a - c) * np.sin(angle) + (b - d) * np.cos(angle) + d])
```
