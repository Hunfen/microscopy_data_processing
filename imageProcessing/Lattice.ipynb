```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axis = plt.subplots(figsize = (5, 5))
area = np.linspace(-5, 5, 11)
area__ =np.linspace(-4.5, 4.5, 10)
va = np.array([1, 0])
vb = np.array([0, 1])
for i in area:
    for j in area:
        v = i * va + j * vb
        handler1 = axis.plot(v[0], v[1], '.', color = 'red')
va_ = va * 3 - vb * 1
vb_ = va * 1 + vb * 2
for i in area:
    for j in area:
        v_ = va_ * i + vb_ * j
        if (v_[0] ** 2 + v_[1] ** 2) <= 26:
            axis.plot(v_[0], v_[1], '.', ms = 20, color = 'Green')
        else:
            continue

for i in area__:
    for j in area__:
        v = i * va + j * vb
        axis.plot(v[0], v[1], '.', color = 'blue')

# Legend
# axis.legend((handler1), ('Se'), bbox_to_anchor=(1.2, 1),loc = 0)


axis.axis('off')
fig.tight_layout();
```
