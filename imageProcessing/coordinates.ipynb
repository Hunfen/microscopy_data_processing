```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker
import skimage
from skimage.io import imread
from skimage import filters
import ipywidgets as widgets
from mpl_toolkits.axisartist.axislines import SubplotZero
```

```python
x = [*np.linspace(-2, -1, 9), *np.linspace(-1, 1, 17), *np.linspace(1, 2, 9)]
y = [*np.ones(9), *np.zeros(17), *np.ones(9)]
x_cos = np.linspace(-1, 1, 201)
y_cos_0 = 0.125 * np.cos(x_cos * np.pi * 0.5) + 0.1
y_cos_1 = 0.07 * np.cos(x_cos * np.pi * 2) + 0.2


fig, axis = plt.subplots()

axis.spines['left'].set_position('zero')
axis.spines['bottom'].set_position('zero')

for direction in ['left', 'bottom', 'right', 'top']:
    axis.spines[direction].set_color('none')

axis.set_xticks([]) 
axis.set_yticks([]) 
axis.arrow(0, -0.2, 0, 1.4, head_width = 0.1, head_length = 0.05, color = 'black')
axis.arrow(-2.1, 0, 4.2, 0, head_width = 0.05, head_length = 0.1, color = 'black')

axis.plot(x, y, ls = '-', lw = 3, color = 'black')
axis.plot(x_cos, y_cos_0, color = 'red')
axis.plot(x_cos, y_cos_1, color = 'blue')

fig.tight_layout();
```
