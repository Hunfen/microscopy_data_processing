```python
import numpy as np
import matplotlib.pyplot as plt
import os, string
from math import sqrt
```

```python
bake_file = '/Users/hunfen/OneDrive/General Files/ゼミー/20201014ベーク.txt'
```

```python
vac_data = np.genfromtxt(bake_file)
np.reshape(vac_data,vac_data.shape)
```

```python
x = []
y1 = []
y2 = []
for i in range(len(vac_data)):
    x.append(vac_data[i][0])
    y1.append(vac_data[i][1])
    y2.append(vac_data[i][2])
```

```python
fig, axis = plt.subplots(figsize = (5, 10/(1+sqrt(5))))
axis.plot(x, y1, '.-', label = 'Load Lock')
axis.plot(x, y2, '.-', label = 'Main Chamber')
axis.set_yscale('log')
# axis.set_xscale('log')
axis.set_xlabel('Time (mins)', font = 'Calibri', fontsize = 14)
axis.set_ylabel('Pressure (Torr)', font = 'Calibri', fontsize = 14)

# inset
""" inset_handle = axis.inset_axes([0.3, 0.3, 0.5, 1/(1+sqrt(5))])
inset_handle.plot(x[6:], y1[6:])
inset_handle.plot(x[6:], y2[6:]) """
# inset_handle.set_ax

""" axis[1].plot(x[6:], y1[6:], '.-', label = 'Load Lock')
axis[1].plot(x[6:], y2[6:], '.-', label = 'Main Chamber')
axis[1].set_yscale('log')
axis[1].set_xscale('log')
axis[1].set_xlabel('Time (min)', font = 'Calibri', fontsize = 14)
axis[1].set_ylabel('Pressure (Torr)', font = 'Calibri', fontsize = 14) """

axis.legend()
fig.tight_layout();
```
